# Avec Mod√®le MLP (Mutlilayer perceptron)

```python
# S√©lectionner les colonnes num√©riques pour la normalisation
numeric_cols = ['Total goals', 'Total Assists', 'Shots per game',
                'Key passes per game', 'Dribbles per game', 'Fouled per game',
                'Offsides per game', 'Dispossessed per game', 'Bad control per game',
                'Saison', 'Dribble', 'Start Application', 'Remplacement Application',
                'Tackles per game', 'Interceptions per game', 'Fouls per game', 'Clearances per game',
                'Dribbled past per game', 'Outfield blocks per game', 'Pass success percentage',
                'Aerials won per game', 'Passes per game', 'Crosses per game', 'Long balls per game',
                'Through balls per game']

# Normaliser les colonnes num√©riques
scaler = StandardScaler()
df_merged[numeric_cols] = scaler.fit_transform(df_merged[numeric_cols])
```

```python
# S√©lectionner les colonnes d'entr√©e et de sortie
X_cols = ['Minutes', 'Total Assists', 'Shots per game',
          'Key passes per game', 'Dribbles per game', 'Fouled per game',
          'Offsides per game', 'Dispossessed per game', 'Bad control per game',
          'Saison', 'Dribble', 'Start Application', 'Remplacement Application',
          'Tackles per game', 'Interceptions per game', 'Fouls per game', 'Clearances per game',
          'Dribbled past per game', 'Outfield blocks per game', 'Pass success percentage',
          'Aerials won per game', 'Passes per game', 'Crosses per game', 'Long balls per game',
          'Through balls per game']

y_col = ['Total goals']

# S√©parer les donn√©es en ensembles d'entra√Ænement et de test
train_data, test_data = train_test_split(df_merged, test_size=0.2, random_state=42)

# Pr√©parer les donn√©es d'entr√©e et de sortie pour l'entra√Ænement du mod√®le
X_train = train_data[X_cols]
y_train = train_data[y_col]

X_test = test_data[X_cols]
y_test = test_data[y_col]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPRegressor

# D√©finir les hyperparam√®tres √† tester
parameters = {
    'hidden_layer_sizes': [(50,), (100,), (100,100)],
    'activation': ['relu', 'logistic'],
    'learning_rate_init': [0.001, 0.01, 0.1],
}

# Cr√©er le MLPRegressor
mlp = MLPRegressor()

# Utiliser GridSearchCV pour trouver les meilleurs hyperparam√®tres
grid_search = GridSearchCV(mlp, parameters, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Afficher les meilleurs hyperparam√®tres et score
print("Best hyperparameters: ", grid_search.best_params_)
print("Best score: ", -grid_search.best_score_)

# Entra√Æner le MLPRegressor avec les meilleurs hyperparam√®tres
mlp = MLPRegressor(hidden_layer_sizes=grid_search.best_params_['hidden_layer_sizes'], 
                   activation=grid_search.best_params_['activation'],
                   learning_rate_init=grid_search.best_params_['learning_rate_init'], solver='adam', max_iter=1000, random_state=42)
mlp.fit(X_train, y_train)

# Faire des pr√©dictions sur les donn√©es de test
y_pred = mlp.predict(X_test)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Cr√©ation de l'objet MLPRegressor
mlp = MLPRegressor(hidden_layer_sizes=(300,300), activation='relu', solver='adam', max_iter=1000, random_state=42)

# Entra√Ænement du mod√®le sur les donn√©es d'entra√Ænement
mlp.fit(X_train, y_train)

# Pr√©dictions sur les donn√©es de test
y_pred = mlp.predict(X_test)

# Evaluation des performances du mod√®le
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("R^2:", r2)
```

```python
from sklearn.multioutput import MultiOutputRegressor

# Entra√Æner un mod√®le pour toutes les colonnes num√©riques (√† l'exception de "Age", "Born" et "Saison")
y_cols = [col for col in numeric_cols if col not in ["Age", "Born", "Saison"]]
y_train_multi = train_data[y_cols]

mlp_multi = MLPRegressor(hidden_layer_sizes=(300, 300), activation='relu', solver='adam', max_iter=1000, random_state=42)
multi_output_regressor = MultiOutputRegressor(mlp_multi).fit(X_train, y_train_multi)
```

```python
def predict_player_performance(player_name, model, scaler, df_merged, df_merged_init):
    player_data = df_merged[df_merged['Name'] == player_name].sort_values(by='Age')
    player_data_init = df_merged_init[df_merged_init['Name'] == player_name].sort_values(by='Age')
    annees = [2024, 2025, 2026]

    for year in annees:
        last_season = player_data.iloc[-1].copy()
        last_season['Saison'] = year
        last_season['Age'] += 1
        player_data = player_data.append(last_season, ignore_index=True)

    # Normaliser les donn√©es
    player_data_normalized = player_data.copy()
    player_data_normalized[numeric_cols] = scaler.transform(player_data_normalized[numeric_cols])

    # Pr√©dictions pour chaque colonne num√©rique
    X_player = player_data_normalized[X_cols]
    y_pred_player = model.predict(X_player)
    
    last_three_rows = player_data_init.tail(4).iloc[:-1]
    # Ajouter les colonnes "Predicted" pour chaque colonne y
    for idx, column in enumerate(y_cols):
        player_data[f"Predicted {column}"] = y_pred_player[:, idx]

    # Si les 3 derni√®res lignes de chaque colonne sont √©gales √† 0, les pr√©dictions seront √©galement nulles
    if (last_three_rows[y_cols] == 0).all(axis=None):
        for year in annees:
            for column in y_cols:
                player_data.loc[player_data['Saison'] == year, column] = 0

        # Mettre √† jour les colonnes "Predicted" avec des valeurs de 0
        for column in y_cols:
            player_data[f"Predicted {column}"] = 0
    else:
        # Ajouter la valeur pr√©dite √† la valeur de l'ann√©e pr√©c√©dente pour chaque colonne num√©rique
        for idx in range(player_data.shape[0] - 1):
            if player_data.iloc[idx + 1]['Saison'] in annees:
                for column in y_cols:
                    player_data.loc[idx + 1, column] = player_data.loc[idx, column] + player_data.loc[idx + 1, f"Predicted {column}"]
    
    # Ajouter les lignes pour les saisons 2024, 2025 et 2026 √† player_data_init
    new_rows = player_data[player_data['Saison'].isin(annees)].copy()
    player_data_init = player_data_init.append(new_rows, ignore_index=True)

    # Calculer la somme des valeurs de la ligne pr√©c√©dente et des valeurs inscrites dans les lignes
    # pour les saisons 2024, 2025 et 2026 (√† l'exception des colonnes "Age", "Born", "Saison" et des colonnes en string)
    for idx in range(player_data_init.shape[0] - 1):
        if player_data_init.iloc[idx + 1]['Saison'] in annees:
            for column in player_data_init.columns:
                if column not in ['Age', 'Born', 'Saison'] and player_data_init[column].dtype != 'object':
                    player_data_init.loc[idx + 1, column] = player_data_init.loc[idx, column] + player_data_init.loc[idx + 1, column]
    # Mettre toutes les valeurs en valeur absolue
    player_data_init[numeric_cols] = player_data_init[numeric_cols].abs()
    
    # Si les 3 derni√®res lignes de chaque colonne sont √©gales √† 0, les pr√©dictions seront √©galement nulles
    last_three_rows = player_data_init.tail(4).iloc[:-1]
    for column in numeric_cols:
        if (last_three_rows[column] == 0).all() or last_three_rows[column].isna().any():
            player_data_init.loc[player_data_init['Saison'].isin(annees), column] = 0
    # Supprimer toutes les colonnes qui commencent par "Predicted"
    player_data_init = player_data_init.drop(columns=[col for col in player_data_init.columns if col.startswith("Predicted")])
    player_data_init = player_data_init.drop('Rating', axis=1)

    return player_data_init

# Utiliser la fonction pour pr√©dire les performances d'un joueur sp√©cifique
player_name = "Lionel Messi"
predicted_performance = predict_player_performance(player_name, multi_output_regressor, scaler, df_merged, df_merged_init)

# Cr√©er un DataFrame avec le nom du joueur et les pr√©dictions
predicted_performance_name = f"{player_name.replace(' ', '_')}_predicted"
globals()[predicted_performance_name] = predicted_performance
print(globals()[predicted_performance_name])
```

```python
def get_predicted_performance(player_name: str) -> pd.DataFrame:
    # Utiliser la fonction pour pr√©dire les performances d'un joueur sp√©cifique
    predicted_performance = predict_player_performance(player_name, multi_output_regressor, scaler, df_merged, df_merged_init)

    # Cr√©er un DataFrame avec le nom du joueur et les pr√©dictions
    predicted_performance_name = f"{player_name.replace(' ', '_')}_predicted"
    globals()[predicted_performance_name] = predicted_performance
    return globals()[predicted_performance_name]
```
