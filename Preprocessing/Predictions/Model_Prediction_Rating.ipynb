```python
#Store a list of all the players
nameList = merged_rating_players.Name.unique()

#create a dataframe just to store the mean values of player rating's 
df_mean = pd.DataFrame(columns=['Name', 'Mean', 'Growth'])
df_mean
```

```python
#Calculating career growth and rating avergae of every player (Feature engineering)
for name in nameList:
    mean = merged_rating_players.loc[merged_rating_players.Name == name]['Rating'].mean()
    length = len(merged_rating_players.loc[merged_rating_players.Name == name]['Rating'])
    growth = merged_rating_players.loc[merged_rating_players.Name == name]['Rating'].tolist()[length-1] - merged_rating_players.loc[merged_rating_players.Name == name]['Rating'].tolist()[0]
    df_mean.at[name, 'Name'] = name
    df_mean.at[name, 'Mean'] = mean
    df_mean.at[name, 'Growth'] = growth
    
#Merging all the data together   
merged_rating_players = pd.merge(merged_rating_players, df_mean, on=['Name'])

#Converting column type
merged_rating_players['Mean'] = merged_rating_players['Mean'].astype(np.float64)
merged_rating_players
```

# Linear Regression 

```python
#Forecasting using linear regression

from statistics import mean 
from sklearn import linear_model 

from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
from sklearn import model_selection
from math import sqrt

accuracy_MSE = []
accuracy_RMSE = []
accuracy_MAE = []
accuracy_R2 = []

#For every season using previous season data to forecast current season rating
for season in range(2011,2022):
    
    training_set = df_transformedPlayer[df_transformedPlayer['Saison'] == season-1]
    testing_set = df_transformedPlayer[df_transformedPlayer['Saison'] == season]

    X_train = training_set.drop(['Rating', 'Name', 'Saison'], axis=1)
    y_train = training_set['Rating'].values
    X_test = testing_set.drop(['Rating', 'Name', 'Saison'], axis=1)
    y_test = testing_set['Rating'].values

    lin_reg = linear_model.LinearRegression(n_jobs=-1) 
    
    lin_reg.fit(X_train, y_train)

    #using k-folds cross validation for calculating forecast accuracy
    scores = cross_val_score(lin_reg, X_test, y_test, cv=10)
    accuracy_R2.append(scores.mean())
    
    y_pred = lin_reg.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = sqrt(mse)
    mae = median_absolute_error(y_test, y_pred)
    
    accuracy_MSE.append(mse)
    accuracy_RMSE.append(rmse)
    accuracy_MAE.append(mae)
     
print('Average accuracy for forecasting player ratings between 2010-2022:')
print("MSE: " + "{:.2f}".format(mean(accuracy_MSE)))
print("RMSE: " + "{:.2f}".format(mean(accuracy_RMSE)))
print("MAE: " + "{:.2f}".format(mean(accuracy_MAE)))
print("R2: " + "{:.2f}".format(mean(accuracy_R2)))
```

```python
#Forecasting player ratings between 2019-2025

for season in range(2019, 2025):
    
    previousPlayers = merged_rating_players.loc[merged_rating_players.Saison == season].copy()
    currentPlayers = merged_rating_players.loc[merged_rating_players.Saison == season+1].copy()
    futurePlayers = currentPlayers.copy()

    futurePlayers['Saison'] = currentPlayers['Saison'].values+1
    futurePlayers['Age'] = currentPlayers['Age'].values+1

    futurePlayers.drop(['Rating'], axis=1)

    li = []
    li.append(previousPlayers)
    li.append(currentPlayers)
    li.append(futurePlayers)

    df_combined = pd.concat(li)
    
    df_combined = df_combined.sort_values(['Saison', 'Name'])

    df_transformedPlayer = df_combined.copy()
    df_transformedPlayer['Last_Season_Rating'] = df_transformedPlayer.groupby(['Name'])['Rating'].shift()
    df_transformedPlayer['Last_Season_Diff'] = df_transformedPlayer.groupby('Name')['Last_Season_Rating'].transform(Series.diff)
    df_transformedPlayer = df_transformedPlayer.dropna()
    
    df_transformedPlayer = df_transformedPlayer.loc[df_transformedPlayer['Saison'] == season+2]
    
    currentSeasonDf = df_transformedPlayer.drop(['Rating', 'Name', 'Saison'], axis=1)
    y_pred = lin_reg.predict(currentSeasonDf)

    #Attaching the predicted values to the original DF so that the cycle can continue 
    #~until the forecasting range comes to an end 
    ratings = ["{:.2f}".format(value) for value in y_pred.tolist()]
    df_transformedPlayer['Rating'] = ratings
    df_transformedPlayer['Rating'] = y_pred.round(2)
    df_transformedPlayer = df_transformedPlayer[['Name', 'Saison', 'Rating', 'Age', 'Mean', 'Growth']]
    merged_rating_players = merged_rating_players.append(df_transformedPlayer, ignore_index=True)
    merged_rating_players.drop_duplicates(subset=['Name', 'Saison'], keep='last', inplace=True)  # ajout de cette ligne
    merged_rating_players = merged_rating_players[merged_rating_players['Rating'].notnull()].copy()
    merged_rating_players['Rating'] = merged_rating_players['Rating'].astype(float)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np
from statistics import mean 
from math import sqrt

# Définition des modèles à tester
models = [
    DecisionTreeRegressor(random_state=0),
    Ridge(alpha=1.0),
    SVR(kernel='linear', C=1.0),
    SVR(kernel='rbf', C=1.0, gamma='scale'),
    RandomForestRegressor(n_estimators=100, random_state=0),
    KNeighborsRegressor(n_neighbors=5)
]
accuracy_MSE = []
accuracy_RMSE = []
accuracy_MAE = []
accuracy_R2 = []

# Stockage des résultats pour chaque modèle
results = {}

for season in range(2010,2022):
    # Préparation des données d'entraînement et de test
    X = df_transformedPlayer_tree.drop(['Rating', 'Name', 'Saison'], axis=1)
    y = df_transformedPlayer_tree['Rating']
           
    training_set = df_transformedPlayer_tree[df_transformedPlayer_tree['Saison'] == season-1]
    testing_set = df_transformedPlayer_tree[df_transformedPlayer_tree['Saison'] == season]

    X_train = training_set.drop(['Rating', 'Name', 'Saison'], axis=1)
    y_train = training_set['Rating'].values
    X_test = testing_set.drop(['Rating', 'Name', 'Saison'], axis=1)
    y_test = testing_set['Rating'].values

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    
for model in models:
        # Initialisation et entraînement du modèle
        reg = model.fit(X_train, y_train)

        # Prédiction sur les données de test
        y_pred = model.predict(X_test)

        # Using k-folds cross validation for calculating forecast accuracy
        scores = cross_val_score(model, X_test, y_test, cv=10)
        accuracy_R2.append(scores.mean())

        mse = mean_squared_error(y_test, y_pred)
        rmse = sqrt(mse)
        mae = median_absolute_error(y_test, y_pred)
        r2_scor = r2_score(y_test, y_pred)

        accuracy_MSE.append(mse)
        accuracy_RMSE.append(rmse)
        accuracy_MAE.append(mae)

        model_name = type(model).__name__
        if model_name not in results:
            results[model_name] = {
                "MSE": [],
                "RMSE": [],
                "MAE": [],
                "R2": [],
                "R2_score":[]
            }

        results[model_name]["MSE"].append(mse)
        results[model_name]["RMSE"].append(rmse)
        results[model_name]["MAE"].append(mae)
        results[model_name]["R2_score"].append(r2_scor)

print('Average accuracy for forecasting player ratings between 2010-2022:')
for model_name in results:
    print(model_name)
    print("MSE: " + "{:.2f}".format(mean(results[model_name]["MSE"])))
    print("RMSE: " + "{:.2f}".format(mean(results[model_name]["RMSE"])))
    print("MAE: " + "{:.2f}".format(mean(results[model_name]["MAE"])))
    print("R2: " + "{:.2f}".format(mean(results[model_name]["R2_score"])))
```

Les deux morceaux de code effectuent des tâches différentes et ne peuvent donc pas être comparés directement. Le premier code utilise la régression linéaire pour prévoir les notes des joueurs pour chaque saison et évalue la performance en utilisant les mêmes métriques sur les données de test. Le deuxième code utilise un modèle d'arbre de décision pour prédire les notes de joueur et évalue la performance du modèle à l'aide de trois métriques : MSE, MAE et R2. Les deux approches sont différentes et ont leurs avantages et leurs inconvénients.

La première approche utilise une régression linéaire (linear_model.LinearRegression) pour prédire les notes des joueurs. Cette méthode est simple et facile à interpréter, mais peut être moins précise que des modèles plus complexes dans certains cas. Elle utilise également une validation croisée pour évaluer la performance du modèle et sélectionner les paramètres optimaux.

La deuxième approche utilise un modèle RandomForestRegressor pour prédire les notes des joueurs. Cette méthode peut être utile lorsque la relation entre les variables explicatives et la variable cible est non-linéaire et complexe. Elle utilise également une validation croisée pour évaluer la performance du modèle et sélectionner les paramètres optimaux.

Il est difficile de dire quelle approche est meilleure car cela dépend du contexte et des données spécifiques. Dans l'ensemble, la première approche peut être plus appropriée si la relation entre les variables est non-linéaire et complexe, tandis que la deuxième approche peut être plus appropriée si la relation entre les variables est linéaire et simple.

Si l'on compare les métriques d'erreur entre les différentes approches, on peut voir que le modèle RandomForestRegressor a une erreur moyenne plus faible (MSE : 0.03, MAE : 0.07) que les autres modèles sur les données de test. Cela suggère que le modèle RandomForestRegressor est plus précis dans ses prévisions que les autres modèles pour ces données. En outre, le modèle RandomForestRegressor a également un coefficient de détermination (R2) similaire à celui des autres modèles (R2 : 0.68), ce qui indique que le modèle est capable d'expliquer une grande partie de la variance des données observées. En considérant ces résultats, il est recommandé d'utiliser le modèle RandomForestRegressor pour la prédiction des notes des joueurs pour cette plage de temps (2010-2022).
