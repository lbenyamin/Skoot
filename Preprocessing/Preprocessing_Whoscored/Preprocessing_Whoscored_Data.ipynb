```python
import re

def remove_parentheses(text):
    return re.sub(r'\([^()]*\)', '', text)

texte = "AM(CLR),FW"
texte_sans_parentheses = remove_parentheses(texte)
print(texte_sans_parentheses)
```

```python
def replace_position(position):
    position = position.strip()  # Enlever les espaces autour de la position
    if "D" in position:
        if "M" in position:
            if "FW" in position:
                position = "Defender, Midfielder, Forward"
            else:
                position = "Defender, Midfielder"
        elif "FW" in position:
            position = "Defender, Forward"
        else:
            position = "Defender"
    elif "M" in position:
        if "FW" in position:
            position = "Midfielder, Forward"
        else:
            position = "Midfielder"
    elif "FW" in position:
        position = "Forward"
    elif "GK" in position:
        position = "Goalkeeper"
    else:
        position = position
    return position

for df in df_list:
    df['Position'].replace(['midfielder'], 'Midfielder', regex=False, inplace=True)
    df_offensive['Position'].replace(['forward', ' Forward  '], 'Forward', regex=False, inplace=True)
    df["Position"] = df["Position"].apply(replace_position)
```

### Apps

```python
def clean_apps(df_list):
    for df in df_list:
        if "Apps" in df.columns:
            # Créer deux nouvelles colonnes pour stocker les valeurs de départ et de remplacement
            df["Start Application"] = df["Apps"].str.extract(r"^(\d+)")
            df["Remplacement Application"] = df["Apps"].str.extract(r"\((\d+)\)$").fillna(0)

            # Transformer les colonnes en int
            df["Start Application"] = pd.to_numeric(df["Start Application"], errors="coerce")
            df["Remplacement Application"] = pd.to_numeric(df["Remplacement Application"], errors="coerce")
            # Supprimer la colonne originale
            df.drop(columns=["Apps"], inplace=True)

    return df_list
```

```python
def convert_to_int(df_list, exclude_cols=[]):
    for df in df_list:
        for col in df.columns:
            if col not in exclude_cols:
                try:
                    # Vérifier si toutes les valeurs décimales de la colonne sont des zéros
                    if all(pd.to_numeric(df[col], errors='coerce').fillna(0) % 1 == 0):
                        df[col] = pd.to_numeric(df[col], downcast='integer')
                except ValueError:
                    pass
```

### Différencier les positions selon les bases de données
##### Mettre les "Midfielder" et "Foward" seulement dans Passing et aussi dans Offensive
##### Mettre les "Midfielder" et "Defensive", "Goalkeeper" seulement dans Defensive
##### Mettre summary dans chacun des database

```python
def drop_rows_with_many_nan(df, threshold):
    # Compter le nombre de NaN par ligne
    nan_count = df.isna().sum(axis=1)
    
    # Sélectionner les lignes qui ont moins de NaN que la limite
    index_to_keep = nan_count[nan_count <= threshold].index
    
    # Retourner le dataframe sans les lignes qui ont trop de NaN
    return df.loc[index_to_keep]
```

```python
# Supprimer les lignes avec plus de 16 NaN, choisis car 8 colonnes sont sure d'être écrite.
# On décide de garder 72,2% des colonnes pleines 
df_merged = drop_rows_with_many_nan(df_merged, threshold=16)
```
