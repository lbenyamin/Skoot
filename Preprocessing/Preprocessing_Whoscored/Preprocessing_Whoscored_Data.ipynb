```python
def replace_position(position):
    position = position.strip()  # Enlever les espaces autour de la position
    positions = position.split(",")  # Diviser la position en une liste de positions
    new_positions = []
    for pos in positions:
        if pos == 'M,FW' or pos == 'AM,FW' or pos =='AM':
            new_positions.append('Midfielder, Forward')
        elif pos == 'M':
            new_positions.append('Midfielder')
        elif pos == 'D,DMC,M' or pos == 'D,DMC' or pos == 'D,M':
            new_positions.append('Defender, Midfielder')
        elif pos == 'D': 
            new_positions.append('Defender')
        elif pos == 'GK':
            new_positions.append('Goalkeeper')
        elif pos == 'FW':
            new_positions.append('Forward')
        elif pos == 'D,FW': 
            new_positions.append('Defender, Forward')
        elif pos == 'D,M,FW': 
            new_positions.append('Defender, Midflieder, Forward')
        elif pos == 'DMC':
            new_positions.append('Defender, Midfielder')
        else:
            new_positions.append(pos)
    new_positions = list(set(new_positions))
    return ', '.join(sorted(new_positions))
```

```python
def clean_positions(df):
    df['Position'] = df['Position'].apply(lambda x: x.strip())
    df['Position'].replace(['midfielder'], 'Midfielder', regex=False, inplace=True)
    df['Position'].replace(['forward'], 'Forward', regex=False, inplace=True)
    df['Position'] = df['Position'].apply(replace_position)
    df['Position'].replace(['Defender, Midfielder, Midfielder', 'Defender, Defender, Midfielder', 'Defender, Defender, Midfielder, Midfielder'], 'Defender, Midfielder', regex=False, inplace=True)
    df['Position'].replace(['Forward, Midfielder', 'Forward, Midfielder, Forward', 'Defender, Midfielder'], 'Midfielder, Forward', regex=False, inplace=True)
    df['Position'].replace(['Defender, Forward, Midfielder'], 'Defender, Midfielder, Forward', regex=False, inplace=True)
    positions = df['Position'].unique()
    print(f"Positions in {df.name}: {positions}\n")
    
for df in df_list:
    df.name = df.columns[0]  # set the name of the dataframe as the name of the column
    clean_positions(df)
```

### Apps

```python
def clean_apps(df_list):
    for df in df_list:
        if "Apps" in df.columns:
            # Créer deux nouvelles colonnes pour stocker les valeurs de départ et de remplacement
            df["Start Application"] = df["Apps"].str.extract(r"^(\d+)")
            df["Remplacement Application"] = df["Apps"].str.extract(r"\((\d+)\)$").fillna(0)

            # Transformer les colonnes en int
            df["Start Application"] = pd.to_numeric(df["Start Application"], errors="coerce")
            df["Remplacement Application"] = pd.to_numeric(df["Remplacement Application"], errors="coerce")
            # Supprimer la colonne originale
            df.drop(columns=["Apps"], inplace=True)

    return df_list
```

```python
def convert_to_int(df_list, exclude_cols=[]):
    for df in df_list:
        for col in df.columns:
            if col not in exclude_cols:
                try:
                    # Vérifier si toutes les valeurs décimales de la colonne sont des zéros
                    if all(pd.to_numeric(df[col], errors='coerce').fillna(0) % 1 == 0):
                        df[col] = pd.to_numeric(df[col], downcast='integer')
                except ValueError:
                    pass
```

### Différencier les positions selon les bases de données
##### Mettre les "Midfielder" et "Foward" seulement dans Passing et aussi dans Offensive
##### Mettre les "Midfielder" et "Defensive", "Goalkeeper" seulement dans Defensive
##### Mettre summary dans chacun des database

```python
def merge_dataframes(dataframes):
    """
    Fusionne plusieurs dataframes en utilisant les colonnes qui se retrouvent dans chaque dataframe
    en tant que clés de fusion.

    :param dataframes: une liste de dataframes à fusionner.
    :return: le dataframe fusionné.
    """
    df_merged = dataframes[0]

    for i in range(1, len(dataframes)):
        df_merged = pd.merge(df_merged, dataframes[i], on=list(set(df_merged.columns) & set(dataframes[i].columns)), how="outer")

    return df_merged
```

```python
def create_position_databases(df):
    # Liste des positions à garder pour chaque base de données
    midfield_positions = ['Midfielder', 'Defender, Midfielder, Forward', 'Midfielder, Forward']
    forward_positions = ['Forward', 'Defender, Forward', 'Midfielder, Forward']
    defensive_positions = ['Defender', 'Defender, Midfielder, Forward', 'Defender, Forward']
    goalkeeper_positions = ['Goalkeeper']
    
   
    # Filtrer les positions et créer les bases de données correspondantes
    df_midfield = df[df['Position'].isin(midfield_positions)]
    df_midfield = merge_dataframes([df_midfield,df_passing])
    df_midfield = merge_dataframes([df_midfield,df_offensive])
    df_midfield = df_midfield.drop('Total goals', axis=1)
    df_midfield = df_midfield[df_midfield['Position'].isin(midfield_positions)]
    
    df_forward = df[df['Position'].isin(forward_positions)]
    df_forward = merge_dataframes([df_forward,df_passing])
    df_forward = merge_dataframes([df_midfield,df_offensive])
    df_forward = df_forward.drop('Total goals', axis=1)
    df_forward = df_forward[df_forward['Position'].isin(forward_positions)]
    
    df_defense = df[df['Position'].isin(defensive_positions)]
    df_defense = merge_dataframes([df_defense,df_passing])
    df_defense = merge_dataframes([df_defense,df_defensive])
    df_defense = df_defense[df_defense['Position'].isin(defensive_positions)]
    
    df_goalkeeper = df[df['Position'].isin(goalkeeper_positions)]
    df_goalkeeper = merge_dataframes([df_goalkeeper,df_defensive])
    df_goalkeeper = df_goalkeeper[df_goalkeeper['Position'].isin(goalkeeper_positions)]
    
    # Enregistrer les bases de données dans des fichiers CSV
    df_midfield.to_csv('midfield_players.csv', index=False)
    df_forward.to_csv('forward_players.csv', index=False)
    df_defensive.to_csv('defensive_players.csv', index=False)
    df_goalkeeper.to_csv('goalkeeper_players.csv', index=False)
```
