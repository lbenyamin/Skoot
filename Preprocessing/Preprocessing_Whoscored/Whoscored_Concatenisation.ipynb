# Concatenate Data

By Offensive, Defensive, Passing and Summary

### **Github Data**




```python
import pandas as pd
import glob
```

```python
all_files = glob.glob("*.csv")
```

```python
offensive_data = []
defensive_data = []
passing_data = []
summary_data = []
for filename in all_files:
    df = pd.read_csv(filename)
    league_name = filename.split('-')[0]  # Récupérer le nom de la ligue sans le préfixe
    saison_name = filename.split("-")[2]
    df['League'] = league_name
    df['Saison'] = saison_name
        # si le fichier contient "Offensive" dans son nom
    if "Offensive" in filename:
            offensive_data.append(df)
    # si le fichier contient "Defensive" dans son nom
    elif "Defensive" in filename:
            # lisez les données dans le fichier et stockez-les dans le dictionnaire
            #data = file.read()
            defensive_data.append(df)
    # si le fichier contient "Passing" dans son nom
    elif "Passing" in filename:
            # lisez les données dans le fichier et stockez-les dans le dictionnaire

            passing_data.append(df)
    # si le fichier contient "Summary" dans son nom
    elif "Summary" in filename:
            # lisez les données dans le fichier et stockez-les dans le dictionnaire
            #data = file.read()

            summary_data.append(df)
    #dfs.append(df)
offensive_data_concatenated = pd.concat(offensive_data, ignore_index= True)
defensive_data_concatenated = pd.concat(defensive_data, ignore_index= True)
passing_data_concatenated = pd.concat(passing_data, ignore_index= True)
summary_data_concatenated = pd.concat(summary_data, ignore_index= True)
```

```python
offensive_data_concatenated.to_csv("Offensive_Whoscored_Data.csv", index=False)
defensive_data_concatenated.to_csv("Defensive_Whoscored_Data.csv", index=False)
passing_data_concatenated.to_csv("Passing_Whoscored_Data.csv", index=False)
summary_data_concatenated.to_csv("Summary_Whoscored_Data.csv", index=False)
```
